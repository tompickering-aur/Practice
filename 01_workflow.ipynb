# 4. How will I integrate (or not) AI

AI tools like ChatGPT, Claude, and GitHub Copilot are increasingly part of data workflows. There is no single right answer — the key is being intentional about how you use them.


**Option 1: No AI — old school coding**

You write all your code yourself. This is a great choice if you want to deepen your understanding of Python and data analysis. You will make more mistakes, but you will also learn more.


**Option 2: AI as a coding assistant**

You stay in control of your notebook, but use AI to help with specific pieces: debugging an error, writing a function, explaining unfamiliar syntax. You read, understand, and consciously decide what to copy into your notebook. This is the recommended starting point for most students.


**Option 3: AI as a collaborator**

You describe what you want to do in plain language and ask AI to generate larger chunks of code or structure. You still review and adapt the output. This works well once you have enough background to evaluate what the AI produces — if you can't tell whether the code is correct, you are not ready for this option yet.


**Option 4: AI with file and data access (advanced)**

Tools like Claude Code or Cursor can be given direct access to your files and repository. This allows for a much more integrated workflow, but requires caution: never give AI access to restricted, sensitive, or proprietary data. You generally need a paid account for this level of integration.


**Key ideas, regardless of option:**

- Always understand what your code does. AI can generate errors just as easily as correct code.
- Document AI-assisted decisions in your notebook, just as you would document your own reasoning.
- AI is particularly useful for: boilerplate code, debugging, reformatting data, writing explanations.
- AI is less reliable for: domain-specific judgment calls, checking statistical validity, interpreting results in context.


**A note on academic integrity**
Check your course or institution's policy on AI use. When in doubt, disclose. Treating AI as a tool you are responsible for — rather than an author — is always the right framing.
